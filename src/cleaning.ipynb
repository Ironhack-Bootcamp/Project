```python
file_path = 'C:/Users/Felix/Documents/Ironhack/Project/data/attacks.csv'

# Try reading the file with a different encoding
data = pd.read_csv(file_path, encoding='latin1')
df = pd.DataFrame(data)
```

### Renaming the columns

```python
df.columns = [i.lower().replace(" ", "_") for i in df.columns]
df.rename(columns={'sex_': 'sex', 'species_': 'species', 'fatal_(y/n)': 'fatal'}, inplace=True)
df
```

### Deleting unnecessary columns 

```python
columnas_a_eliminar = ["unnamed:_22", "unnamed:_23", "href", "href_formula", "original_order", "pdf", "case_number.1", "case_number", "case_number.2", "time"]
df.drop(columns=columnas_a_eliminar, inplace=True)
```

### Deleting empty rows

```python
df.dropna(how="all", inplace=True)
```

#### Cleaning column: "type"

```python
df['type'] = df['type'].str.lower()

df['type'] = df['type'].str.extract(r'(unprovoked|provoked)')

df['type'] = df['type'].fillna('-1')
```

#### Cleaning column: "age"

```python
df['age'] = df['age'].str.replace(r'\D', '', regex=True)

df['age'] = df['age'].replace('', -1)

df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(-1).astype(int)

df['age'] = df['age'].apply(lambda x: x if 0 <= x <= 100 else -1)
```

```python
df['age'].value_counts()
```

#### Cleaning column: "year"

```python
df['year'] = df['year'].fillna('').astype(str)

df['year'] = df['year'].str.extract(r'(\d{4})')

df['year'] = df['year'].fillna('-1')

df['year'] = df['year'].astype(int)
```

```python
df['year'].value_counts()
```

#### Cleaning column: "date"

```python
def buscar_fecha(texto):
    patron = r'.*?(\d{2}-[a-zA-Z]+-\d{4}).*'
    coincidencias = re.findall(patron, texto)
    if coincidencias:
        return coincidencias[0]
    return '-1'

df['date'] = df['date'].apply(buscar_fecha)

df['date'] = df['date'].fillna('-1')
```

```python
df['date'].value_counts()
```

#### Cleaning column: "country"

```python
df['country'] = df['country'].str.replace(r'[^\w\s,]', '', regex=True)

df['country'].fillna(-1)
```

```python
gc = geonamescache.GeonamesCache()

# Obtener una lista de nombres de países
pais_lista = [pais['name'].lower() for pais in gc.get_countries().values()]
pais_lista.extend(['usa'])
```

```python
paises = pais_lista
```

```python
df['country'] = df['country'].str.lower()
df['country'] = df['country'].apply(lambda x: -1 if x not in paises else x)
```

#### Cleaning column: "activity"

```python
df['activity'].value_counts().head(25)
```

```python
df['activity'] = df['activity'].str.lower()

# Lista de actividades extraídas del texto
data = """Surfing
Swimming
Fishing
Spearfishing
Bathing
Wading
Diving
Standing
Snorkeling
Scuba diving
Body boarding
Body surfing
Kayaking
Paddle
Kayak
Treading water
Pearl diving
Fell overboard
Free diving
Boogie boarding
Windsurfing
Shark fishing"""

def unify_words(match):
    return match.group(1).lower()

# Aplica la función de unificación usando regex para unificar palabras iguales
df['activity'] = df['activity'].str.replace(r'(\b\w+\b)(?=\s+\1)', unify_words, regex=True)

activity_counts = df['activity'].value_counts()
```

#### Cleaning column: "sex"

```python
df['sex'] = np.where(df['sex'].isin(['M', 'F']), df['sex'], -1)
```

#### Cleaning column: "fatal"

```python
df['fatal'] = np.where(df['fatal'].isin(['N', 'Y']), df['fatal'], -1)
```

#### Cleaning: "species"

```python
df['species'] = df['species'].str.lower()

valid_species = df['species'].apply(lambda x: re.sub(r"[0-9'`[].*$]", '', str(x)))
valid_species = valid_species.str.replace('invalid', '').str.replace('shark involvement prior to death was not confirmed', '')

shark_names = [
    "white shark", "hammerhead shark", "tiger shark", "bull shark", 
    "nurse shark", "mako shark", "thresher shark", "blue shark", 
    "lemon shark", "whale shark", "blacktip shark", "silky shark", 
    "goblin shark", "dogfish shark", "sandbar shark", "megamouth shark", 
    "porbeagle shark", "greenland shark", "zebra shark", "wobbegong shark", 
]

valid_species = valid_species.apply(lambda x: next((name for name in shark_names if name in x), 'unknown'))

df['species'] = valid_species
```
