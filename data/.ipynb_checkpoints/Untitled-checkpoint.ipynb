```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
file_path = 'C:/Users/Felix/Documents/Ironhack/Project/data/attacks.csv'

# Try reading the file with a different encoding
data = pd.read_csv(file_path, encoding='latin1')
df = pd.DataFrame(data)
```

```python
print(data.shape)
data.head()
```

### Renaming the columns

```python
df.columns = [i.lower().replace(" ", "_") for i in df.columns]
df.rename(columns={'sex_': 'sex', 'species_': 'species', 'fatal_(y/n)': 'fatal'}, inplace=True)
df
```

### Deleting empty rows

```python
df.dropna(how="all", inplace=True)
```

### Cleaning a bit age column

```python
df['age'] = df['age'].str.replace(r'\D', '', regex=True)

df['age'].replace('', pd.NA, inplace=True)

df['age'] = df['age'].str.extract('(\d+)').astype(float)

df['age'] = df['age'].fillna(0).astype(int)

df = df[df['age']<=100]
```

### Cleaning a bit year column

```python
df['year'] = df['year'].fillna('').astype(str)

# Utilizar str.extract() para extraer el año
df['year'] = df['year'].str.extract(r'(\d{4})')

# Rellenar los valores nulos en 'year' con ceros (puedes elegir otro valor si lo prefieres)
df['year'] = df['year'].fillna('0')

# Convertir la columna 'year' a enteros
df['year'] = df['year'].astype(int)

# Contar la frecuencia de cada año
conteo_year = df['year'].value_counts()
```

### Creating a new column with values from: date and year

```python
# Unificar las fechas en una nueva columna "Fecha Unificada"
df['unified_date'] = pd.NaT  # Crear una columna vacía de fechas
date_formats = ['%Y.%m.%d', '%d-%b-%Y']
for format in date_formats:
    
    df['unified_date'] = df['unified_date'].combine_first(pd.to_datetime(df['date'], format=format, errors='coerce'))
    
df['unified_date'] = df['unified_date'].combine_first(pd.to_datetime(df['year'], format='%Y', errors='coerce'))
```

### Deleting unnecessary columns 

```python
columnas_a_eliminar = ["unnamed:_22", "unnamed:_23", "href", "href_formula", "original_order", "pdf", "case_number.1", "case_number", "date", "year", "case_number.2"]
df.drop(columns=columnas_a_eliminar, inplace=True)
```

### Deleting duplicated values

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['country'] = df['country'].fillna('unknown')
```

```python
missing_values = df.isna().sum()
df.columns
```

```python
cols_cat = ['country', 'area','activity', 'name', 'sex_', 'species_', 'fatal_(y/n)']
```

```python
# Crear subtramas
fig, ax = plt.subplots(nrows=10, ncols=1, figsize=(10, 30))
fig.subplots_adjust(hspace=1)

for i, col in enumerate(cols_cat):
    sns.countplot(x=col, data=df, ax=ax[i])
    
    # Define las ubicaciones de las marcas de los ejes (ajusta según tus necesidades)
    ax[i].set_xticks(range(len(df[col].unique())))
    
    # Establecer etiquetas de los ejes
    ax[i].set_xticklabels(df[col].unique(), rotation=30)

# Mostrar los gráficos
plt.show()
```
